```python
## Importing Necessary Libraries
import requests
import json
from pathlib import Path
from datetime import datetime
```

```python
url = "https://workflowhub.eu/workflows.json?filter[workflow_type]=galaxy"
```

```python
## Seting Up Directories
output_dir = Path("../data/raw/workflowhub")
output_dir.mkdir(parents=True, exist_ok=True)
```

```python
today = datetime.now().strftime("%Y-%m-%d")
output_file = output_dir / f"workflowhub_galaxy_workflows_{today}.json"

response = requests.get(url, timeout=60)
response.raise_for_status()
```

```python
data = response.json()

with open(output_file, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

print("Saved:", output_file.resolve())
```
